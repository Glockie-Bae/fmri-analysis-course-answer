```python
# Load libraries

import nibabel as nib
from nilearn import plotting, image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load data

fmri_path = 'data/sub-01_ses-1_task-rest_bold_space-MNI152NLin2009cAsym_preproc.nii.gz'
fmri = image.load_img(fmri_path)
```

```python
# Print dimensions and type of data
print(type(fmri))
print(fmri.shape)
```

```python
# 2. 选择一个时间点 (比如第0个volume)
from nilearn.image import index_img
fmri_vol0 = index_img(fmri, 0)  # 取第0个时间点的3D volume
```

```python
# 3. 用 plot_epi 绘制这个volume
plotting.plot_epi(fmri_vol0, title="fMRI Volume 0")
plotting.show()  # 在脚本里记得调用 show()
```

```python
# Convert nibabel file to numpy array (get_data)
fmri_array = fmri.get_fdata()

# Check type and dimensions
print(type(fmri_array))
print(fmri_array.shape)
```
