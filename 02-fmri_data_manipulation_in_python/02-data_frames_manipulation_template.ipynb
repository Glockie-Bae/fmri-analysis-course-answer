## 1. Load & plot confounds variables 

**Confounds** (or *nuisance regressors*) are variables representing fluctuations with a potential non-neuronal origin. 

Confounding variables calculated by [fMRIPrep](https://fmriprep.readthedocs.io/en/stable/outputs.html#confounds) are stored separately for each subject, session and run in TSV files - one column for each confound variable. 

```python
# Load data

confounds_path = "data/sub-01_ses-1_task-rest_bold_confounds.tsv"

# 读取 confounds TSV 文件
confounds = pd.read_csv(confounds_path, sep='\t')

# 查看前5行
print(confounds.head())
```

```python
# Print column names
print(confounds.shape)
print(confounds.columns)
```

```python
# Plot mean timeseries from cerebrospinal fluid (CSF) and white matter 
# 找到 CSF 和 WhiteMatter 相关列
csf_columns = [col for col in confounds.columns if "CSF" in col]
wm_columns  = [col for col in confounds.columns if "WhiteMatter" in col]

# 计算平均时间序列
csf_mean = confounds[csf_columns].mean(axis=1)
wm_mean  = confounds[wm_columns].mean(axis=1)

# 绘图
plt.figure(figsize=(10, 4))
plt.plot(csf_mean, label="CSF mean")
plt.plot(wm_mean, label="White Matter mean")
plt.xlabel("Time (TRs)")
plt.ylabel("Signal")
plt.title("Mean CSF and White Matter Confounds")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot cerebrospinal fluid (CSF) and white matter timeseries on a scatterplot


# 绘制散点图
plt.figure(figsize=(6, 6))
plt.scatter(csf_mean, wm_mean, alpha=0.6, color='blue')
plt.xlabel("CSF mean signal")
plt.ylabel("White Matter mean signal")
plt.title("CSF vs White Matter Scatterplot")
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Plot cerebrospinal fluid (CSF) and white matter timeseries on a scatterplot (use seaborn joint plot)
import seaborn as sns
# 放入 DataFrame 方便 Seaborn 绘图
df_plot = pd.DataFrame({
    "CSF": csf_mean,
    "WhiteMatter": wm_mean
})

# 使用 Seaborn jointplot 绘制散点图 + 边缘直方图
sns.jointplot(data=df_plot, x="CSF", y="WhiteMatter", kind="scatter", color="blue", alpha=0.6)

plt.suptitle("CSF vs White Matter Joint Scatterplot", y=1.02)  # 给图加标题
plt.show()
```
